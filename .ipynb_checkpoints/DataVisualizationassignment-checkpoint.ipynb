```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
# Set seed for reproducibility


import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Set seed for reproducibility
np.random.seed(0)

# Generate dataset of random numbers
n_points = 1000
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, n_points)

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True)
plt.title('Distribution of Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
# Boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Boxplot of Numerical Variable')
plt.show()

# Q-Q Plot
plt.figure(figsize=(10, 6))
stats.probplot(data, plot=plt)
plt.title('Q-Q Plot of Numerical Variable')
plt.show()

```

```python
#3.Create a dataset representing categories and their corresponding values. Compare different categories  based on numerical values.
# Create dataset
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 35, 20]

# Create DataFrame
df = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# Print DataFrame
print(df)

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Category'], df['Value'])
plt.title('Comparison of Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Pie chart
plt.figure(figsize=(10, 6))
plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%')
plt.title('Distribution of Categories')
plt.show()
# Sort values in descending order
df_sorted = df.sort_values(by='Value', ascending=False)

# Print sorted DataFrame
print(df_sorted)
```

```python
#4.Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories
# Generate dataset
np.random.seed(0)
categories = ['A', 'B', 'C', 'D']
values = np.random.normal(0, 1, 100)
category_labels = np.random.choice(categories, 100)
data = pd.DataFrame({
    'Category': category_labels,
    'Value': values
})

# Print dataset
print(data.head())

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=data)
plt.title('Distribution of Values Across Categories')
plt.show()

# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Value', data=data)
plt.title('Distribution of Values Across Categories')
plt.show()

# Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Category', y='Value', data=data)
plt.title('Distribution of Values Across Categories')
plt.show()

# Bar Plot (mean values)
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=data, estimator=np.mean)
plt.title('Mean Values Across Categories')
plt.show()
```

```python
#5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

 #Set seed for reproducibility
np.random.seed(0)

# Generate synthetic dataset with correlated features
n_samples = 1000
n_features = 5

# Define correlation structure
corr_matrix = np.array([
    [1.0, 0.8, 0.4, 0.2, 0.1],
    [0.8, 1.0, 0.7, 0.5, 0.3],
    [0.4, 0.7, 1.0, 0.9, 0.6],
    [0.2, 0.5, 0.9, 1.0, 0.8],
    [0.1, 0.3, 0.6, 0.8, 1.0]
])

# Generate random data with specified correlation
data = np.random.multivariate_normal(mean=[0, 0, 0, 0, 0], cov=corr_matrix, size=n_samples)

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])

# Calculate correlation matrix
corr_df = df.corr()

# Visualize correlation matrix using heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()

```

```python
#1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space
import plotly.express as px
import pandas as pd
import numpy as np
# Set seed for reproducibility
np.random.seed(30)

# Generate dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Generate 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z')

# Update plot layout
fig.update_layout(
    title='3D Distribution of Data Points',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    width=800,
    height=600
)

# Show plot
fig.show()
```

```python
#2.Using the Student Grades, create a violin plot to display the distribution of scores across different grade  categories.
# Set seed for reproducibility
np.random.seed(15)

# Generate dataset
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Generate violin plot
fig = px.violin(df, x='Grade', y='Score', box=True, hover_data=df.columns)

# Update plot layout
fig.update_layout(
    title='Distribution of Scores Across Grades',
    xaxis_title='Grade',
    yaxis_title='Score'
)

# Show plot
fig.show()
```

```python
#3.Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# Set seed for reproducibility

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(20)

# Generate dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Pivot DataFrame for heatmap
df_pivot = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc=np.mean)

# Generate heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_pivot, annot=True, cmap='coolwarm', square=True, fmt='.0f')

# Update plot layout
plt.title('Variation in Sales Across Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
#4.Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(under root x2+y2)
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create meshgrid
x, y = np.meshgrid(x, y)

# Calculate z values
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten data for DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Generate 3D surface plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Viridis')])
# Update plot layout
fig.update_layout(
    title='3D Surface Plot of z = sin((x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    width=800,
    height=600
)

# Show plot
fig.show()
```

```python
#5.. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
# Set seed for reproducibility
np.random.seed(25)

# Generate dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create bubble chart
fig = px.scatter(df, x='GDP', y='Population', 
                 size='Population', hover_name='Country',
                 title='Country Population and GDP')
# Update plot layout
fig.update_layout(
    xaxis_title='GDP (Billions)',
    yaxis_title='Population (Millions)'
)

# Show plot
fig.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, output_notebook, show

x = np.linspace(0, 10, 100)
y = np.sin(x)

output_notebook()
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)

show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(0)
x = np.random.normal(0, 10, 100)
y = np.random.normal(0, 10, 100)
sizes = np.random.randint(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue', 'yellow', 'purple'], 100)
# Create ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, size=sizes, color=colors))

# Create Bokeh plot
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter plot with varying sizes and colors
p.scatter('x', 'y', size='size', fill_color='color', source=source)

# Show plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create Bokeh plot
p = figure(title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           x_range=fruits)
# Add bar chart
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Show plot
show(p)
```

```python
#4.Create a Bokeh histogram to visualize the distribution of the given data.
# Generate random data
np.random.seed(0)
data_hist = np.random.randn(1000)

# Create histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create Bokeh plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency',
           x_range=[edges[0], edges[-1]])
# Add histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='skyblue', line_color='black')

# Show plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256

# Generate random data
np.random.seed(0)
data_heatmap = np.random.rand(10, 10)
# Create x and y coordinates
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create Bokeh plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=[0, 1], y_range=[0, 1], tools="hover")

# Add heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette=Viridis256)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)

```
